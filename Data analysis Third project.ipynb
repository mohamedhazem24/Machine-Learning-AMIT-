```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv('voice.csv')
df
```

```python
df.replace({'male':1,'female':0},inplace=True)
```

```python
df_male=df[df['label']==1]
```

```python
df_female=df[df['label']==0]
```

```python
df_female.describe()
```

```python
df_male.describe()
```

## we concluded from the two above describe table:
### 1-that he used the same number of man as the women to give us an unbiased data 
### 2- females have higher mean of frequency than men which is resonable cause females have sound of higher pitch
### 3- the average variance of males is higher than the average variance of females because most males have hard voices but you can find some a moderate hardness voice but most of the females have soft voice
### 4-the spectral entropy mean of females approches more to zero than males does and that is a proof that females have more pure tone in their voices and can preform better in singing

```python
plt.figure(figsize=(10,10))
plt.hist(df_male['meanfreq'],color='cyan',bins=30,histtype='step')
plt.hist(df_female['meanfreq'],color='pink',bins=30,histtype='step')
plt.xlabel('meanfreq')
plt.show()
```

### we noticed that there are frequency too high that female reached but male had not and thats make us conclude that vocal cords of males are thicker than females and that makes female vocal cords are easier to vibrate per second (Hz) than males do

```python
plt.figure(figsize=(10,10))
plt.hist(df_male['IQR'],color='cyan',bins=30,histtype='step')
plt.hist(df_female['IQR'],color='pink',bins=30,histtype='step')
plt.xlabel('meanfreq')
plt.show()
```

### we can notice that the  range of changing of frequency of males is more than females
